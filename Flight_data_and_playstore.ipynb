```python
'''#Q3 the max min and range of price'''

# Calculate maximum price
max_price = df['Price'].max()
# Calculate minimum price
min_price = df['Price'].min()

# Calculate range of prices
price_range = max_price - min_price

print(f"Maximum Price: ${max_price:.2f}")
print(f"Minimum Price: ${min_price:.2f}")
print(f"Range of Prices: ${price_range:.2f}")
```

```python
'''Q6:You are working for a travel agency, and your boss has asked you to analyze the Flight Price dataset to identify the peak travel season. What features would you analyze to identify the peak season, and how would you present your findings to your boss?'''

```

```python
'''Answer:
Patterns and seasonal variations in pricing trends are found by tracking airline fares over time. We can observe how airline costs vary over time and how they connect to travel demand by using a time series visualisation. This gives us insights into when individuals travel the most and when costs tend to rise, enabling us to identify peak travel seasons when prices are at their highest.'''
```

```python
df['Date_of_Journey'] = pd.to_datetime(df['Date_of_Journey'])
```

```python
df['Month'] = df['Date_of_Journey'].dt.month
df['Year'] = df['Date_of_Journey'].dt.year
```

```python
avg_prices_by_month = df.groupby(['Year', 'Month'])['Price'].mean()
plt.figure(figsize=(12, 6))
avg_prices_by_month.plot(marker='x')
plt.title('Average Flight Prices Over Time')
plt.xlabel('Month')
plt.ylabel('Average Price')
plt.grid(True)
plt.xticks(rotation=45)
```

```python
'''Q7 : You are a data analyst for a flight booking website, and you have been asked to analyze the Flight Price dataset to identify any trends in flight prices. What features would you analyze to identify these trends, and what visualizations would you use to present your findings to your team?'''
```

```python
'''Answer 7 : Price trends can be learned a lot by tracking changes in travel costs over time, including weekly, monthly, and seasonal fluctuations. Furthermore, recognising price differences between various routes and destinations is essential for determining well-liked travel routes and locations with distinct price points. Seasonal variations in airfare for particular routes can be easily observed by utilising heatmaps to visualise average flight fares across several months and routes. With the help of this thorough study, stakeholders in the aviation industry may maximise revenue generation and customer happiness by making well-informed decisions on pricing strategies and resource allocation.

'''
```

```python
heatmap_data = df.pivot_table(index='Month', columns='Destination', values='Price', aggfunc='mean')
plt.figure(figsize=(10, 6))
sns.heatmap(heatmap_data, cmap='Set2', annot=True)
plt.title('Heatmap of Flight Prices by Month and Destination')
plt.xlabel('Destination')
plt.ylabel('Month')
```

```python
def convert_size(size):
  if 'M' in size:
    return int(float(size.replace('M',''))*1000)
  elif 'K' in size:
    return float(size.replace('k',''))
  elif 'Varies with device' in size:
    return np.nan
  else:
    return np.nan
```

```python
df['Size']=df['Size'].apply(convert_size)
```

```python
'''Q16. A mobile app development company wants to analyze the Google Playstore dataset to identify the most successful app developers. What features would you analyze to make recommendations to the company, and what data visualizations would you use to present your findings?'''
```

```python
'''Let's look at each app individually, as the data does not include the app developers. With the paid apps, we can determine the revenue just by counting the installs.
'''
```

```python
# Cleanning 'Installs' column to remove non-numeric values
df['Installs'] = df['Installs'].str.replace(',', '').str.replace('+', '')
```

```python
# Converting 'Installs' to numeric format
df['Installs'] = pd.to_numeric(df['Installs'], errors='coerce')
```

```python
newDf = df.drop_duplicates(subset='App', keep='first')
revenueApp = newDf['Installs'] * newDf['Price']
top_10_revenue = revenueApp[revenueApp.values != 0].sort_values(ascending = False).head(10)
top_10_revenue.index = newDf.loc[top_10_revenue.index, 'App']
```

```python
top_10_revenue
```

```python
sns.barplot(y=top_10_revenue.index, x=top_10_revenue.values)
plt.xlabel('Revenue Generated')
plt.show()
```
